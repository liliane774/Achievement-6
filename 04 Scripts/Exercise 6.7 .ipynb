```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulated dataset based on the provided structure

data = {
    'Region': ['North America', 'North America', 'Europe', 'Asia', 'Asia', 'Africa', 'North America', 'Europe'],
    'Country': ['USA', 'Canada', 'Germany', 'India', 'China', 'Kenya', 'USA', 'France'],
    'Year': [2020, 2020, 2021, 2021, 2022, 2022, 2023, 2023],
    'AvgTemperature': [15.2, 10.8, 12.5, 28.3, 18.9, 24.6, 16.5, 13.2],
    'GDP_Growth': [2.3, 1.8, 2.5, 6.1, 4.9, 5.2, 2.1, 1.9],
    'Transportation_Demand': [1200, 950, 1100, 1800, 1600, 700, 1250, 1050],
    'Infrastructure_Risk_Score': [4, 3, 5, 7, 6, 8, 5, 4],
    'Seasonal_Peak': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No']
}
df = pd.DataFrame(data)
```

```python
# Basic Statistics
print("Basic Statistics:")
print(df.describe())
```

```python
# Visualization 1: Scatter Plot of AvgTemperature vs Transportation_Demand
df_path = '/Users/ryumugil/Downloads/Achievement 6 Analysis/03 Analysis/Visualizations/demand_by_peak.png'
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='AvgTemperature', y='Transportation_Demand', hue='Region', size='Infrastructure_Risk_Score')
plt.title('Temperature vs Transportation Demand by Region')
plt.xlabel('Average Temperature (C)')
plt.ylabel('Transportation Demand (Units)')
plt.grid(True)
#plt.show()
plt.savefig(df_path)
```

```python
# Visualization 2: Box Plot of GDP_Growth by Seasonal_Peak
df_path = '/Users/ryumugil/Downloads/Achievement 6 Analysis/03 Analysis/Visualizations/gdp_growth_over_time.png'
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Seasonal_Peak', y='GDP_Growth')
plt.title('GDP Growth by Seasonal Peak')
plt.xlabel('Seasonal Peak')
plt.ylabel('GDP Growth (%)')
plt.grid(True)
#plt.show()
plt.savefig(df_path)
```

```python
# Visualize the relationship
df_path = '/Users/ryumugil/Downloads/Achievement 6 Analysis/03 Analysis/Visualizations/temp_demand_regression'
plt.figure(figsize=(10, 6))
#sns.scatterplot(data=df, x='AvgTemperature (C)', y='Transportation_Demand (Units)', hue='Region', s=100)
plt.plot(df['AvgTemperature'], color='red', linewidth=2, label='Regression Line')
plt.title('Transportation Demand vs. Temperature by Region')
plt.xlabel('Average Temperature')
plt.ylabel('Transportation Demand (Units)')
plt.legend()
plt.grid(True)
#plt.show()
plt.savefig(df_path)
```
